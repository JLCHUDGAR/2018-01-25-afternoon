```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import train_test_split
%matplotlib inline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn import tree
```

```python
df = pd.read_csv('adult.csv', header=None, skipinitialspace=True)
```

```python
def country_to_numeric(x):
    if x=='United States':
        return 1
    if x=='Outlying-US(Guam-USVI-etc)':
        return 2
    if x=='Puerto-Rico':
        return 3
    if x=='Ecuador':
        return 4
    if x=='Peru':
        return 5
    if x=='Columbia':
        return 6
    if x=='Honduras':
        return 7
    if x=='Nicaragua':
        return 8
    if x=='Guatemala':
        return 9
    if x=='El-Salvador':
        return 10
    if x=='Greece':
        return 11
    if x=='Holand-Netherlands':
        return 12
    if x=='England':
        return 13
    if x=='France':
        return 14
    if x=='Germany':    
        return 15
    if x=='Poland':
        return 16
    if x=='Italy':
        return 17
    if x=='Ireland':
        return 18
    if x=='Scotland':
        return 19
    if x=='Hungary':    
        return 20
    if x=='Portugal':    
        return 21
    if x=='Yugoslavia':    
        return 22
    if x=='India':
        return 23
    if x=='Canada':    
        return 24
    if x=='Mexico':    
        return 25
    if x=='Trinadad&Tobago':
        return 26
    if x=='Haiti':    
        return 27
    if x=='Dominican-Republic':    
        return 28
    if x=='Jamaica':    
        return 29
    if x=='Cuba':    
        return 30
    if x=='Laos':
        return 31
    if x=='Thailand':    
        return 32
    if x=='Cambodia':    
        return 33
    if x=='Hong-Kong':    
        return 34
    if x=='Taiwan':    
        return 35
    if x=='Japan':
        return 36
    if x=='Vietnam':    
        return 37
    if x=='China':    
        return 38
    if x=='South Korea':    
        return 39
    if x=='Philippines':    
        return 40
    if x=='Iran':    
        return 41
    if x=='?':    
        return 42
    else:
        return 0

def predict_to_numeric(z):
    if z=='<=50K':
        return 0
    else:
        return 1
    
def race_to_numeric(r):
    if r=='White':
        return 1
    if r=='Black':
        return 2
    if r=='Asian-Pac-Islander':
        return 3
    if r=='Amer-Indian-Eskimo':
        return 4
    else:
        return 0
    
def sex_to_numeric(s):
    if s=='Male':
        return 1
    else:
        return 0

def workclass_to_numeric(w):
    if w=='Self-emp-not-inc':
        return 1
    if w=='Self-emp-inc':
        return 2
    if w=='Private':
        return 3
    if w=='Federal-gov':
        return 4
    if w=='State-gov':
        return 5
    if w=='Local-gov':
        return 6
    else:
        return 0
    
def maritalstatus_to_numeric(m):
    if m=='Never-married':
        return 1
    if m=='Divorced':
        return 2
    if m=='Seperated':
        return 3
    if m=='Widowed':
        return 4
    if m=='Married-AF-spouse':
        return 5
    if m=='Married-civ-spouse':
        return 6
    if m=='Married-spouse-absent':
        return 7
    else:
        return 0
    
def education_to_numeric(e):
    if e=='Preschool':
        return 1
    if e=='1st-4th':
        return 1
    if e=='5th-6th':
        return 1
    if e=='7th-8th':
        return 1
    if e=='9th':
        return 2
    if e=='10th':
        return 2
    if e=='11th':
        return 2
    if e=='12th':
        return 2
    if e=='HS-grad':
        return 3
    if e=='Some-college':
        return 4
    if e=='Assoc-voc':
        return 5
    if e=='Assoc-acdm':
        return 5
    if e=='Prof-school':
        return 6
    if e=='Bachelors':
        return 6
    if e=='Masters':
        return 7
    if e=='Doctorate':
        return 7
    else:
        return 0

def relationship_to_numeric(r):
    if r=='Husband':
        return 1
    if r=='Not-in-family':
        return 2
    if r=='Other-relative':
        return 3
    if r=='Own-child':
        return 4
    if r=='Unmarried':
        return 5
    if r=='Wife':
        return 6
    else:
        return 0

def occupation_to_numeric(x):
    if x=='Adm-clerical':
        return 1
    if x=='Armed-Forces':
        return 2
    if x=='Craft-repair':
        return 3
    if x=='Exec-managerial':
        return 4
    if x=='Farming-fishing':
        return 5
    if x=='Handlers-cleaners':
        return 6
    if x=='Machine-op-inspct':
        return 7
    if x=='Other-service':
        return 8
    if x=='Priv-house-serv':
        return 9
    if x=='Prof-specialty':
        return 10
    if x=='Protective-serv':
        return 11
    if x=='Sales':
        return 12
    if x=='Tech-support':
        return 13
    if x=='Transport-moving':
        return 14
    else:
        return 0
```

```python
df['workscore'] = df['workclass'].apply(workclass_to_numeric)
df['maritalscore'] = df['maritalstatus'].apply(maritalstatus_to_numeric)
df['ocupationscore'] = df['occupation'].apply(occupation_to_numeric)
df['educatiobuckets'] = df['education'].apply(education_to_numeric)
df['relationshipscore'] = df['relationship'].apply(relationship_to_numeric)
df['racescore'] = df['race'].apply(race_to_numeric)
df['male'] = df['sex'].apply(sex_to_numeric)
df['countryscore'] = df['native-country'].apply(country_to_numeric)
df['Over50k'] = df['predictmean'].apply(predict_to_numeric)
df
```

```python
dfnumeric= df.drop(["workclass", "education", "maritalstatus", "occupation", "relationship", "race", "sex","native-country", "predictmean"], axis = 1)
```

```python
dfnumeric
```

```python
y = dfnumeric.Over50k
X = dfnumeric.drop(["Over50k"], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100).fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
cv = KFold(n_splits=10)
```

```python
for depth in range(1, 20):
    model = tree.DecisionTreeClassifier(max_depth=depth)
    train_scores, test_scores = [], []
    for train_i, test_i in cv.split(X):
        Xr, yr, Xt, yt = X.loc[train_i], y.loc[train_i], X.loc[test_i], y.loc[test_i]
        model.fit(Xr, yr)
        test_scores.append(model.score(Xt, yt))
        train_scores.append(model.score(Xr, yr))
    print('depth:', depth, 'test scores:', sum(test_scores)/len(test_scores), 'train scores:', sum(train_scores)/len(train_scores))
```

```python
cv = KFold(n_splits=10)
for estimators in range(100, 500, 100):
    model = RandomForestClassifier(n_estimators=estimators, max_depth=9)
    scores = []
    for train_i, test_i in cv.split(X):
        Xr, yr, Xt, yt = X.loc[train_i], y.loc[train_i], X.loc[test_i], y.loc[test_i]
        model.fit(Xr, yr)
        scores.append(model.score(Xt, yt))
        train_scores.append(model.score(Xr, yr))
    print('estimators:' , estimators, 'scores:', sum(scores)/len(scores))
```

```python

final_model = RandomForestClassifier(n_estimators=400, max_depth=9)

```

```python
final_model.fit(X, y)
```
